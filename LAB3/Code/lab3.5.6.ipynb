# Lab3.5 Sử dụng multiindex và groupby

```python
import pandas as pd
```

```python
#1 Đọc file stocks1.csv vào DataFrame stocks1
file_path = r"D:\17A3KHDL\LAB3\Data\stocks1.csv"
stocks1 = pd.read_csv(file_path)
print(stocks1)
```

```python
#1, Đọc file stocks2.csv vào DataFrame stocks2 
stocks2 = pd.read_csv(r"D:\17A3KHDL\LAB3\Data\stocks2.csv")
print(stocks2)
```

```python
# Gộp dữ liệu 
stocks = pd.concat([stocks1, stocks2])
stocks
```

```python
# 1. Tạo MultiIndex cho DataFrame bằng cột 'date' và 'symbol'
stocks.set_index(['date', 'symbol'], inplace=True)
stocks
```

```python
# 2. Sử dụng GroupBy để tính giá trung bình và volume trung bình
tinh_gia_tb = stocks.groupby(['date', 'symbol']).mean()
tinh_gia_tb
```

```python
# 3. Sắp xếp dữ liệu theo ngày và mã chứng khoán
sap_xep_du_lieu = tinh_gia_tb.sort_index()
sap_xep_du_lieu
```

```python
#4 Hiển thị kết quả cho 5 ngày đầu tiên
print("\n Kết quả của 5 ngày đầu tiên là:", sap_xep_du_lieu.head())
```

# 3.6 Tạo và phân tích pivot table

```python
#1 Tạo pivot table từ dataframe stocks với date làm chỉ mục, symbol làm cột, và giá trị trung bình của close làm giá trị 
pivot_table = stocks.pivot_table(values= 'close', index = 'date', columns = 'symbol', aggfunc='mean')
pivot_table
```

```python
# 2. Thêm một cột tính tổng volume giao dịch cho mỗi mã chứng khoán (symbol)
volume_sum = stocks.groupby('symbol')['volume'].sum()
volume_sum
```

```python
# 3. Sắp xếp Pivot Table dựa trên tổng volume giao dịch, từ cao xuống thấp
sx_pivot_table = volume_sum.sort_values(ascending=False)
sx_pivot_table
```

```python
# 4. Hiển thị kết quả cho 5 mã chứng khoán có tổng volume giao dịch cao nhất
print(sx_pivot_table.head(5))
```
